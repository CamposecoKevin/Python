```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
sns.barplot(x=["A","B","C"], y= [1,2,3])
plt.show()
```

```python
sns.set(style= "dark", palette= "dark" )
sns.barplot(x=["A","B","C"], y= [1,2,3])
plt.show()
```

```python
tip= sns.load_dataset('tips')
```

```python
tip
```

```python
sns.displot(data=tip, x="total_bill", y = "tip")
plt.show()
```

```python
sns.displot(data=tip, x="total_bill",)
plt.show()
```

```python
sns.displot(data=tip, x="total_bill", y = "tip", hue= "sex")
plt.show()
```

```python
sns.displot(data=tip, x="total_bill", hue= "sex")
plt.show()
```

```python
sns.histplot(data=tip, x="tip", bins=15, cumulative=False, hue= "sex")
plt.show()
```

```python
sns.histplot(data=tip, x="tip", bins=15, cumulative=False, hue= "sex", multiple="stack")
plt.show()
```

```python
sns.histplot(data=tip, x="tip", bins=15, cumulative=False, hue= "sex", multiple="dodge")
plt.show()
```

```python
sns.kdeplot(data= tip, x='tip', hue = "sex", fill = True)
plt.show()
```

```python
sns.countplot(data=tip, x="day", hue = "sex", )
plt.show()
```
